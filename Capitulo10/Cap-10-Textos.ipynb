**Extracción de un carácter**

```python
mensaje = "Hola, Grupo"
caracterPos0 = mensaje[0]
caracterPos4 = mensaje[4]
caracterPosFinal = mensaje[len(mensaje)-1]

print(caracterPos0)
print(caracterPos4)
print(caracterPosFinal)
```

**Extracción de un subtexto**

```python
texto = "Precio: 250"
subTexto1 = texto[0:7]
subTexto2 = texto[8:11]

print(subTexto1)
print(subTexto2)
```

**Búsqueda de un carácter o subtexto**

```python
mensaje = "Hola, Grupo"
posO = mensaje.find("o")
posGrupo = mensaje.find("Grupo")
posZ = mensaje.find("Z")

print(posO)
print(posGrupo)
print(posZ)
```

**Función vs método**

```python
mensaje = "Hola, Grupo"
posH = mensaje.find("H")
print(posH)

edad = 28
pos8 = edad.find(8)
```

**Búsqueda de un carácter o subtexto basado en una posición inicial**

```python
texto = "Precio: 250 - Cantidad: 8"
pos1 = texto.find(":")
pos2 = texto.find(":", pos1+1)
subTextoCantidad = texto[pos2+2:pos2+3]

print(pos1)
print(pos2)
print(subTextoCantidad)
```

**Otros métodos**

```python
mensaje = "Ubicación: provenza"

# retorna el texto en minúscula
mensajeMinus = mensaje.lower()
print(mensajeMinus)

# retorna el texto en mayúscula
mensajeMayus = mensaje.upper()
print(mensajeMayus)

# cuenta cantidad de "a"
conteo = mensaje.count("a")
print(conteo)

# retorna reemplazando "a" por "x"
nuevoMensaje = mensaje.replace("a", "x")
print(nuevoMensaje)
```

**Ciclos con textos**

```python
palabra = input("Ingrese una palabra: ")

i = 0
while(i < len(palabra)):
  print(palabra[i])
  i = i+1
```
