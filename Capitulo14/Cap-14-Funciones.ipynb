**Funciones con retorno**

```python
def calcularMulta(diasDeRetraso):
  if(diasDeRetraso > 7):
    return diasDeRetraso*2
  else:
    return 3

dias = int(input("Ingrese los días de retraso: "))
multa = calcularMulta(dias)
print("Debe pagar "+str(multa)+" dólares de multa")
```

**Calculadora**

```python
def sumar():
  num1 = float(input("Ingrese el primer número: "))
  num2 = float(input("Ingrese el segundo número: "))
  resultado = num1+num2
  print("La suma es: "+str(resultado))

while(True):
  print("----Ingrese----")
  print("1) Para sumar")
  print("0) Para salir")
  accion = int(input())
  if(accion == 1):
    sumar()
  elif (accion == 0):
    print("Chao")
    Break
```

**Acción restar**

```python
def sumar():
  num1 = float(input("Ingrese el primer número: "))
  num2 = float(input("Ingrese el segundo número: "))
  resultado = num1+num2
  print("La suma es: "+str(resultado))

def restar():
  num1 = float(input("Ingrese el primer número: "))
  num2 = float(input("Ingrese el segundo número: "))
  resultado = num1-num2
  print("La resta es: "+str(resultado))

while(True):
  print("----Ingrese----")
  print("1) Para sumar")
  print("2) Para restar")
  print("0) Para salir")
  accion = int(input())
  if(accion == 1):
    sumar()
  elif(accion == 2):
    restar()
  elif (accion == 0):
    print("Chao")
    Break
```
